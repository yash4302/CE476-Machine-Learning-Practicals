```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import arff

%matplotlib inline
```

```python
data = arff.loadarff('credit-g.arff')
df = pd.DataFrame(data[0])
df
```

```python
df.replace(to_replace='?', value=np.NaN)
df
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:, 1:]
x
```

```python
y = df.iloc[:, :1]
y
```

```python
from sklearn.preprocessing import LabelEncoder as le

for column in x.columns:
  if column == 'credit_amount':
    continue
  x[column] = le().fit_transform(x[column])

x
```

```python
y['checking_status'] = le().fit_transform(y['checking_status'])
y
```

```python
y = y.to_numpy()
y
```

```python
from sklearn.preprocessing import OneHotEncoder as oe

x = oe().fit_transform(x).toarray()
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split as test

x_train, x_test, y_train, y_test = test(x, y, test_size=0.10, random_state=10)

print('X_train : ', x_train)
print('X_test : ', x_test)
print('Y_train : ', y_train)
print('Y_test : ', y_test)
```

```python
print('X : ', x.shape)
print('Y : ', y.shape)
print('X_train : ', x_train.shape)
print('X_test : ', x_test.shape)
print('Y_train : ', y_train.shape)
print('Y_test : ', y_test.shape)
```

```python
y_train = y_train.ravel()
y_test = y_test.ravel()

y_train
```

```python
from sklearn.linear_model import LogisticRegression as lr

logist = lr(penalty='l2', solver='lbfgs', max_iter=1000, verbose=1)
model = logist.fit(x_train, y_train)
predictions = logist.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import accuracy_score as acs
from sklearn.metrics import classification_report as cr

print('Confusion Matrix :\n', cm(y_test, predictions))
print('\nAccuracy Score : ', acs(y_test, predictions))
print('\nReport :\n', cr(y_test, predictions))
```
