```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# http://ftp.cs.toronto.edu/pub/neuron/delve/data/tarfiles/comp-activ.tar.gz
df = pd.read_csv('comp-activ.data', sep=' ', ).iloc[:, :27]
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x = df[['lread', 'lwrite', 'scall', 'sread', 'swrite']]
y = df['usr']
```

```python
from sklearn.model_selection import train_test_split as test

x_train, x_test, y_train, y_test = test(x, y, test_size=0.2, random_state=10)
```

```python
print(x.shape)
print(y.shape)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler as sc
from sklearn.preprocessing import Normalizer as nm

scaler = sc().fit(x_train)
scaler_x_train = scaler.transform(x_train)
normalized = nm().fit(x_train)
normalized_x_train = normalized.transform(x_train)

scaler = sc().fit(x_test)
scaler_x_test = scaler.transform(x_test)
normalized = nm().fit(x_test)
normalized_x_test = normalized.transform(x_test)

print('Scalar X-train : ', scaler_x_train, '\n')
print('Scalar X-test : ', scaler_x_test, '\n')
print('Normalized X-train : ', scaler_x_train, '\n')
print('Normalized X-test : ', scaler_x_test, '\n')
```

```python
from sklearn.feature_selection import SelectKBest as best
from sklearn.feature_selection import chi2 as ch

bestFeature = best(score_func=ch, k=3)
fit = bestFeature.fit(x, y)
dfScore = pd.DataFrame(fit.scores_)
dfColumns = pd.DataFrame(x.columns)
featureScores = pd.concat([dfColumns, dfScore], axis=1)
featureScores.columns = ['User', 'Score']
featureScores.nlargest(3, 'Score')
```

```python
bestFeature = best(score_func=ch, k=3)
fit = bestFeature.fit(x_train, y_train)
dfScore = pd.DataFrame(fit.scores_)
dfColumns = pd.DataFrame(x_train.columns)
featureScores = pd.concat([dfColumns, dfScore], axis=1)
featureScores.columns = ['User', 'Score']
featureScores.nlargest(3, 'Score')
```

```python
bestFeature = best(score_func=ch, k=3)
fit = bestFeature.fit(x_test, y_test)
dfScore = pd.DataFrame(fit.scores_)
dfColumns = pd.DataFrame(x_test.columns)
featureScores = pd.concat([dfColumns, dfScore], axis=1)
featureScores.columns = ['User', 'Score']
featureScores.nlargest(3, 'Score')
```
