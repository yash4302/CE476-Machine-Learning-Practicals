```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataframe = pd.read_csv('imdb-videogames.csv', index_col=0)
dataframe
```

```python
new_df = dataframe.dropna()
new_df
```

```python
new_df.replace(to_replace='-',value=np.NaN,inplace=True)
new_df
```

```python
new_df.dropna(inplace=True)
new_df
```

```python
plt.figure(figsize = (14,10))
sns.boxplot(x='year', y='rating', data=new_df)
```

```python
plt.figure(figsize = (14,10))
plt.scatter(new_df['year'], new_df['rating'])
```

```python
plt.figure(figsize = (14,10))
plt.hist(new_df['rating'])
```

```python
corrMatrix = new_df.corr()
corrMatrix
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corrMatrix, annot=True)
```
